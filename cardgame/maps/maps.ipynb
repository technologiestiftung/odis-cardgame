```python
"""
------------------------------------------------------------------------------
GeoJson Markers
------------------------------------------------------------------------------
"""

main_folder = "../maps/prognoseraum_1/"
geojson_data = {
    "type": "FeatureCollection",
    "features": []
}

for prognoseraum in os.listdir(main_folder):
    prognoseraum_path = os.path.join(main_folder, prognoseraum)

    if os.path.isdir(prognoseraum_path):  # Ignore .DS_Store-Dateien
        #iterate over files in prognoseraum
        for filename in os.listdir(prognoseraum_path):
            if filename.endswith(".geojson"):
                with open(os.path.join(prognoseraum_path, filename), "r") as geojson_file:
                    data = json.load(geojson_file)
                    
                    #FeatureCollection
                    if data.get("type") == "FeatureCollection":
                        geojson_data["features"].extend(data["features"])
                    else:
                        print(f"{filename} ist keine gültige FeatureCollection.")
                        
# main_folder = "../maps/prognoseraum_1/"

# for prognoseraum in os.listdir(main_folder):
#     prognoseraum_path = os.path.join(main_folder, prognoseraum)

#     if os.path.isdir(prognoseraum_path): #ignores .DS_store file
#         geojson_data = []

#         #iterate
#         for i in os.listdir(prognoseraum_path):
#             if i.endswith(".geojson"):
#                 with open(os.path.join(prognoseraum_path, i), "r") as geojson_file:
#                     data = json.load(geojson_file)
#                     geojson_data.append(data)

        #create layer
        # layers = []
        # for geojson in geojson_data:
        #     layers.append({
        #         'type': 'geojson',
        #         'data': geojson, 
        #         'format': 'geojson'
        #  })
            
   
        
```
