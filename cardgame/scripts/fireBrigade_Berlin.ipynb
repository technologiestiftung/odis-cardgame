```python
"""
------------------------------------------------------------------------------
Libraries
------------------------------------------------------------------------------
"""

import pandas as pd
import geopandas as gpd

import matplotlib.pyplot as plt
```

```python
"""
------------------------------------------------------------------------------
Import data
------------------------------------------------------------------------------
"""

df = pd.read_csv("../data/raw/BFw_district_area_data_2023.csv", dtype={"district_area_id":str})

df.head()
```

```python
df_pgr = pd.read_csv("../data/raw/PGR.csv", dtype={"PGR_ID":str})

df_pgr = df_pgr[["PGR_ID", "PGR_NAME"]]

df_pgr = df_pgr.rename(columns={"PGR_ID": "pgr_area_id", "PGR_NAME": "PGR_NAME"})

df_pgr.head()
```

```python
#add missing 0 in district_area_id
df["district_area_id"] = df["district_area_id"].str.zfill(6)

df.head(50)
```

```python
df["pgr_area_id"] = df["district_area_id"].astype(str).str[:4]
```

```python
"""
------------------------------------------------------------------------------
mission_count_all:
Dies ist die Gesamtanzahl aller Einsätze, die die Feuerwehr in einem bestimmten 
Zeitraum durchgeführt hat. Es umfasst alle Arten von Einsätzen, also Notfalleinsätze, 
Feuerwehreinsätze, technische Rettungen usw.

mission_count_ems:
Diese Zahl gibt die Anzahl der Notfalleinsätze (Emergency Medical Services, EMS) an. 
EMS-Einsätze umfassen medizinische Notfälle, bei denen die Feuerwehr medizinische Hilfe leisten muss, 
aber nicht unbedingt bei einem Brand oder einer technischen Rettung beteiligt ist.

mission_count_ems_critical:
Dies zeigt die Anzahl der kritischen Notfalleinsätze innerhalb der EMS-Einsätze an. 
Kritische EMS-Einsätze sind besonders dringende medizinische Notfälle, bei denen eine schnelle 
Reaktion erforderlich ist, um das Leben der betroffenen Personen zu retten.

mission_count_ems_critical_cpr:
Diese Zahl bezieht sich auf die Anzahl der kritischen EMS-Einsätze, bei denen eine 
Herz-Lungen-Wiederbelebung (CPR) durchgeführt wurde. Diese Einsätze sind besonders dringend, 
da sie häufig mit einem Herzstillstand oder schwerwiegenden Atemproblemen verbunden sind.

mission_count_fire:
Dies ist die Anzahl der Feuerwehreinsätze. Diese Kategorie umfasst Einsätze, 
bei denen die Feuerwehr tätig wird, um Brände zu bekämpfen oder Brandgefahren zu bewältigen.

mission_count_technical_rescue:
Diese Zahl gibt die Anzahl der technischen Rettungseinsätze an. Technische Rettungen umfassen 
spezielle Rettungsaktionen, die nicht direkt mit Bränden oder medizinischen Notfällen zu tun haben, 
wie z.B. Rettungen aus Höhenlagen, Fahrzeugbergungen oder das Befreien von Personen aus 
eingeklemmten Situationen.

Mission Count EMS Critical Timegoal Computed: 
Anzahl der kritischen Notfalleinsätze, bei denen die Reaktionszeitvorgaben 
eingehalten wurden.

Mission Count Fire Timegoal Computed: 
Anzahl der Feuerwehreinsätze, bei denen die Reaktionszeitvorgaben eingehalten wurden.
------------------------------------------------------------------------------
"""
```

```python
"""
------------------------------------------------------------------------------
Subset
- mission_count_all
- mission_count_ems
- mission_count_ems_critical
- mission_count_fire
- mission_count_technical_rescue
- Mission Count EMS Critical Timegoal Computed
- Mission Count Fire Timegoal Computed
------------------------------------------------------------------------------
"""

list = ["mission_count_all", "mission_count_ems", "mission_count_ems_critical", "mission_count_fire", "mission_count_technical_rescue", "mission_count_ems_critical_timegoal_computed", "mission_count_fire_timegoal_computed"]

df_timegoal = df[["pgr_area_id", "mission_count_all", "mission_count_ems", "mission_count_ems_critical", "mission_count_fire", "mission_count_technical_rescue", "mission_count_ems_critical_timegoal_computed", "mission_count_fire_timegoal_computed"]]
#group by pgr_area_id" and aggregate values
df_timegoal = df_timegoal.groupby("pgr_area_id")[list].sum().reset_index()

df_timegoal.head()


```

```python
"""
------------------------------------------------------------------------------
Merge
------------------------------------------------------------------------------
"""

timegoal_pgr = pd.merge(df_timegoal, df_pgr, on= "pgr_area_id", how="left")

timegoal_pgr.head()
```

```python
timegoal_pgr.isna().sum()
```

```python
timegoal_pgr.dtypes
```

```python
#all mission calls including ems + fire + technical rescue
timegoal_pgr["mission_ems_fire_technical"] = timegoal_pgr["mission_count_ems"] + timegoal_pgr["mission_count_fire"] + timegoal_pgr["mission_count_technical_rescue"]

#add column mission_ems_fire_technical on fourth position
columns = ["pgr_area_id", "PGR_NAME", "mission_count_all", "mission_ems_fire_technical", "mission_count_ems", "mission_count_ems_critical", "mission_count_fire", "mission_count_technical_rescue", "mission_count_ems_critical_timegoal_computed", "mission_count_fire_timegoal_computed"]
timegoal_pgr = timegoal_pgr[columns]

timegoal_pgr.head()
```

```python
"""
------------------------------------------------------------------------------
percent time goal reached
------------------------------------------------------------------------------
"""
#timegoal reached ems critical / all mission counts ems critical
timegoal_pgr["percent_timegoal_ems_critical"] = (timegoal_pgr["mission_count_ems_critical_timegoal_computed"] / timegoal_pgr["mission_count_ems_critical"]).round(3) *100 
#timegoal reached fire / all mission counts fire
timegoal_pgr["percent_timegoal_fire"] = (timegoal_pgr["mission_count_fire_timegoal_computed"] / timegoal_pgr["mission_count_fire"]).round(3) *100 
```
