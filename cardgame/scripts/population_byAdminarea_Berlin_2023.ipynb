```python
"""
------------------------------------------------------------------------------
Libraries
------------------------------------------------------------------------------
"""

import pandas as pd
```

```python
"""
------------------------------------------------------------------------------
Import data
------------------------------------------------------------------------------
"""

df = pd.read_csv("../data/raw/Einwohner_Berlin_2023_nachLOR.csv",delimiter=';',dtype={"Bezirk": str, "Prognoseraum": str, "Bezirksregion": str, "Planungsraum": str, "LOR": str})

df.head(5)
```

```python
df_pgr = pd.read_csv("../data/raw/PGR.csv", dtype={"PGR_ID": str})
df_pgr.head()
```

```python
"""
------------------------------------------------------------------------------
Merge
------------------------------------------------------------------------------
"""

df["PGR_ID"] = df["LOR"].str[:4]

df.head()

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)
pd.set_option("display.width", 1000)


df_merged = pd.merge(df, df_pgr, on= "PGR_ID", how="left")

print(df_merged)
```

```python
#Test
#len(df_merged["pgr_name"].unique())
```

```python
#Removal of spaces and hyphens in "Anzahl" column and conversion to int
df_merged["Anzahl"] = df_merged["Anzahl"].replace({"–": "0"}, regex=True) 
df_merged["Anzahl"] = df_merged["Anzahl"].str.replace(" ", "", regex=True).astype(int)

```

```python
"""
------------------------------------------------------------------------------
Population per admin area
------------------------------------------------------------------------------
"""

df_Einwohner_nachPGR = df_merged.groupby("PGR_NAME")["Anzahl"].sum().reset_index()

df_Einwohner_nachPGR.head()
```
