```python
"""
------------------------------------------------------------------------------
About
The geojson file of the admin areas is used to get the total area in km2
for each.
------------------------------------------------------------------------------
"""
```

```python
"""
------------------------------------------------------------------------------
Libraries
------------------------------------------------------------------------------
"""

import pandas as pd
import geopandas as gpd
```

```python
"""
------------------------------------------------------------------------------
Import data
------------------------------------------------------------------------------
"""

admin_areas = gpd.read_file("../data/raw/Prognoseräume.json")

admin_areas.head()
```

```python
#check crs
admin_areas.crs
```

```python
#convert crs to ESPG:25833
gdf_adminArea = admin_areas.to_crs(epsg=25833)

gdf_adminArea.crs
```

```python
#area for each geometry
gdf_adminArea["area_km2"] = gdf_adminArea.geometry.area /1000000 #m2 in km2

#subset
adminArea = gdf_adminArea[["pgr_id", "pgr_name", "area_km2"]] 

adminArea["area_km2"] = adminArea["area_km2"].round(2)

adminArea.sort_values(by= "area_km2", ascending=False, inplace=True)
adminArea.reset_index(drop=True, inplace=True)

adminArea
```

```python
#csv
adminArea.to_csv("../data/csv/adminAreaKm2.csv", index= False)
```
