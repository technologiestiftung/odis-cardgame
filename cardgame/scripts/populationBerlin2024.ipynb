```python
"""
------------------------------------------------------------------------------
About
In this data analysis, the population in 2024 is determined for each admin area.
------------------------------------------------------------------------------
"""
```

```python
"""
------------------------------------------------------------------------------
Libraries
------------------------------------------------------------------------------
"""

import pandas as pd
```

```python
"""
------------------------------------------------------------------------------
Import data
------------------------------------------------------------------------------
"""

df = pd.read_excel("../data/raw/EinwohnerzahlBerlin2024.xlsx",dtype={"Bezirk": str, "Prognoseraum": str, "Bezirksregion": str, "Planungsraum": str, "LOR": str})

df.head(5)
```

```python
df_pgr = pd.read_csv("../data/raw/PGR.csv", dtype={"PGR_ID": str})
df_pgr.head()
```

```python
"""
------------------------------------------------------------------------------
Merge
------------------------------------------------------------------------------
"""

#add column PGR_ID in df
df["pgr_id"] = df["LOR"].str[:4]

#rename column
df_pgr.rename(columns= {"PGR_ID":"pgr_id", "PGR_NAME":"pgr_name"}, inplace=True)


pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)
pd.set_option("display.width", 1000)

#merge data frames
df_merged = pd.merge(df, df_pgr, on= "pgr_id", how="left")

print(df_merged)
```

```python
"""
------------------------------------------------------------------------------
Population per admin area
------------------------------------------------------------------------------
"""

populationBerlin2024= df_merged.groupby(["pgr_id","pgr_name"])["Einwohnerzahl"].sum().reset_index()

#sort descending
populationBerlin2024.sort_values(by="Einwohnerzahl", ascending=False, inplace=True)
populationBerlin2024.reset_index(drop=True, inplace=True)

#rename column
populationBerlin2024.rename(columns={"Einwohnerzahl":"population"}, inplace=True)
populationBerlin2024
```

```python
#csv
populationBerlin2024.to_csv("../data/csv/populationBerlin2024.csv", index=False)
```
